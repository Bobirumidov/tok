# 4

```python
x = tok_df[["2012", '2017', '2018', "2019",'2020', "2021", "2022", '2023', '2024']]
```

```python
y= tok_df['2015']
```

```python
print(type(x), type(y))
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size = 0.3,
                                                    random_state = 42)
```

```python
# Keling, treningning yangi yo'nalishini ko'rib chiqaylik
print(x_train.shape, y_train.shape)

# sinov namunasi
print(x_test.shape, y_test.shape)
```

# 5

```python
# Biz chiziqli regressiyani sklearn kutubxonasining chiziqli modellar to'plamidan import qilamiz.
from sklearn.linear_model import LinearRegression

#Keling, ushbu klassning ob'ektini yaratamiz va uni model o'zgaruvchisiga yozamiz.
model = LinearRegression()

# modelni o'rgatish
model.fit(x_train, y_train)
```

```python
# Bir nechta mustaqil o'zgaruvchilarga (X) asoslanib, biz uy-joy narxini (y) taxmin qilamiz.
y_pred = model.predict(x_test)

# Keling, bir qator indekslardan foydalangan holda dastlabki beshta qiymatni chop qilaylik
print(y_pred[:])
```

```python
# o'rtacha kvadrat xato funktsiyasini import qiling
from sklearn.metrics import root_mean_squared_error

# taqqoslash
print('Root Mean Squared Error (RMSE):', root_mean_squared_error(y_test, y_pred))
```

```python
# Keling, yana bir ko'rsatkichni ko'rib chiqaylik, aniqlanish koeffitsienti R2
from sklearn.metrics import r2_score
print('R2:', np.round(r2_score(y_test, y_pred), 2))
```

# 6

```python
r2_score(y_test, y_pred), model.score(x_test, y_test)
```
